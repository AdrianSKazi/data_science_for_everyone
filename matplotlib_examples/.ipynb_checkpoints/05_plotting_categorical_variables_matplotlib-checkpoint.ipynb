# Plotting Categorical Variables with matplotlib

```python
import matplotlib.pyplot as plt
```

```python
data = {'Dan':10, 'Kerry':15, 'Bob':2,'Libby':4}
names = data.keys()
values = data.values()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10,5), sharey=True)
ax[0].bar(names, values)
ax[1].scatter(names, values)
ax[2].plot(names, values)
fig.suptitle("Categorical Variables Example");
```

# Plotting Categorical Variables with matplotlib 2

```python
boy = ["bored","happy", "bored","bored","happy","bored"]
girl = ["happy", "happy", "happy","bored","happy", "bored"]
activity = ["breakfast","reading","game","snack","nap", "bathtime"]
```

```python
fig, ax = plt.subplots()
ax.plot(activity, boy, label="boy")
ax.plot(activity, girl, label="girl")
ax.legend()
plt.show()
```
