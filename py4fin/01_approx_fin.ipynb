# Mathematical Tools for Finance
## Approximation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
def create_plot(x, y, styles, labels, axlabels):
    plt.figure(figsize=(10,6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
create_plot([x], [f(x)], ['b'], ['f(x)'], ['x','f(x)'])
```

## Regression

$$min_{\alpha_{1}\cdots\alpha_{D}} \frac{1}{I}\sum_{i=1}^{I}\left(y_{i}-\sum_{d=1}^{D}\alpha_{d}\cdot\beta_{d}(x_{i})\right)$$

```python
res = np.polyfit(x, f(x), deg=1, full=True) # lin reg
```

```python
res # Full results
```

```python
ry = np.polyval(res[0], x) # Eval Reg param
```

```python
create_plot([x,x],[f(x), ry], ['b', 'r.'], ['f(x)','lin_reg'], ['x','f(x)'])
```

```python
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)
```

```python
create_plot([x,x],[f(x), ry], ['b', 'r.'], ['f(x)','d5_reg'], ['x','f(x)'])
```

```python
reg = np.polyfit(x, f(x), deg=7)
ry = np.polyval(reg, x)
```

```python
create_plot([x,x],[f(x), ry], ['b', 'r.'], ['f(x)','d7_reg'], ['x','f(x)'])
```

```python
np.allclose(f(x), ry) # Check func and reg value are same (or at least close)
```

```python
# MSE
np.mean((f(x)- ry)**2)
```

## Individual Basis Functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
def create_plot(x, y, styles, labels, axlabels):
    plt.figure(figsize=(10,6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
mat = np.zeros((3+1, len(x)))
```

```python
mat
```

```python
mat[3, :] = x**3 #cubic
mat[2, :] = x**2
mat[1, :] = x
mat[0, :] = 1 # constant
```

```python
reg = np.linalg.lstsq(mat.T, f(x), rcond=None)[0]
```

```python
reg.round(4) # optimal reg params
```

```python
ry = np.dot(reg, mat)
```

```python
create_plot([x,x], [f(x), ry], ['b','r.'], ['f(x)', 'reg'], ['x', 'f(x)'])
```

```python
mat[3, :] = np.sin(x)
```

```python
reg = np.linalg.lstsq(mat.T, f(x), rcond=None)[0]
```

```python
ry = np.dot(reg, mat)
```

```python
np.allclose(f(x), ry)
```

```python
np.mean((f(x)-ry)**2)
```

```python
create_plot([x,x], [f(x), ry], ['b','r.'], ['f(x)', 'reg'], ['x', 'f(x)'])
```

## Noisy Data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-2 * np.pi, 2 * np.pi, 50)

def f(x):
    return np.sin(x) + 0.5 * x

def create_plot(x, y, styles, labels, axlabels):
    """Simplification function for plotting"""
    plt.figure(figsize=(10,6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
xn = np.linspace(-2 * np.pi, 2 * np.pi, 50) # x values
xn = xn + 0.15 * np.random.standard_normal(len(xn)) # add random noise
yn = f(xn) + 0.25 * np.random.standard_normal(len(xn)) # y values with rand noise
```

```python
reg = np.polyfit(xn, yn, 7)
ry = np.polyval(reg, xn)
```

```python
create_plot([xn,xn], [f(xn), ry], ['b', 'r.'], ['f(x)', 'd7_reg'], ['x', 'f(x)'])
```

## Unsorted Data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
def f(x):
    return np.sin(x) + 0.5 * x

def create_plot(x, y, styles, labels, axlabels):
    """Simplification function for plotting"""
    plt.figure(figsize=(10,6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
xu = np.random.rand(50) * 4 * np.pi - 2 * np.pi
yu = f(xu)
```

```python
reg = np.polyfit(xu, yu, 5)
ry = np.polyval(reg, xu)
```

```python
create_plot([xu, xu], [yu, ry], ['b.', 'ro'], ['f(x)', '5d_reg'], ['x', 'f(x)'])
```

## Multi-Dimensional Data

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def fm(p):
    x, y = p
    return np.sin(x) + 0.25 * x + np.sqrt(y) + 0.05 * y ** 2
```

```python
# X, Y, Z
x = np.linspace(0,10, 20)
y = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x, y) # 2D grid out of 1D arrays
```

```python
Z = fm((X, Y))
x = X.flatten() # 2D to 1D
y = Y.flatten()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection="3d")
surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap="coolwarm", linewidth=0.5, antialiased=True)
```

```python
mat = np.zeros((len(x), 6+1))
mat[:,6] = np.sqrt(y)
mat[:,5] = np.sin(x)
mat[:,4] = y ** 2
mat[:,3] = x ** 2
mat[:,2] = y
mat[:,1] = x
mat[:,0] = 1
```

```python
reg = np.linalg.lstsq(mat, fm((x,y)), rcond=None)[0]
```

```python
RZ = np.dot(mat, reg).reshape((20,20)) # Transforms reg to the grid structure
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection="3d")
surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap="coolwarm", linewidth=0.5, antialiased=True)
wire = ax.plot_wireframe(X, Y, RZ, rstride=2, cstride=2, color="black", linewidth=5, label="regression")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.legend()
fig.colorbar(surf, shrink=0.5, aspect=5);
```
